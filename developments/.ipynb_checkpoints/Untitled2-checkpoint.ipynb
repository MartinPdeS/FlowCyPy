```python
def remove_duplicate_labels(ax):
    """
    Removes duplicate labels from the legend of a given axis.

    Parameters
    ----------
    ax : matplotlib.axes.Axes
        The axis to clean up the legend.
    """
    handles, labels = ax.get_legend_handles_labels()
    unique = dict(zip(labels, handles))  # Preserves order, removes duplicates
    ax.legend(unique.values(), unique.keys())
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt


def generate_data(n_samples=5000, seed=42):
    """
    Return:
      X : shape (n_samples, 8) => [p1, p2, p3, p4, m1, m2, m3, m4]
      Y : shape (n_samples, 4) => [A1, A2, A3, A4] (true amplitudes)
    
    We'll also keep widths random or constant. We'll store them
    internally to compute the measured amplitudes, but we won't
    feed them to the NN (the user can decide).
    """
    np.random.seed(seed)
    
    # 4 pulses per sample
    n_pulses = 3

    # 1) Random positions & widths
    #    For each sample, we get p_j, w_j (or keep them constant if you prefer)
    positions = np.random.uniform(0.0, 1.0, size=(n_samples, n_pulses))
    widths    = np.random.uniform(0.05, 0.05, size=(n_samples, n_pulses))

    # 2) True amplitudes => each A_j is random
    true_amps = np.random.uniform(0.5, 2.0, size=(n_samples, n_pulses))

    # 3) Compute measured amplitude m_j at x = p_j
    #    m_j = sum_{k=1..4} A_k * exp( - (p_j - p_k)^2 / (2 w_k^2) )
    measured_amps = np.zeros_like(true_amps)
    for i in range(n_samples):
        for j in range(n_pulses):
            pj = positions[i, j]
            # sum contributions from all pulses k=1..4
            total = 0.0
            for k in range(n_pulses):
                pk = positions[i, k]
                wk = widths[i, k]
                ak = true_amps[i, k]
                contrib = ak * np.exp( -0.5 * ((pj - pk)/wk)**2 )
                total += contrib
            measured_amps[i, j] = total

    # X => stack positions & measured => shape (n_samples, 8)
    X = np.concatenate([positions, measured_amps], axis=1)
    # Y => the 4 true amplitudes => shape (n_samples, 4)
    Y = true_amps

    return X, Y

##############################################################################
# 2) Build a Supervised NN to Regress from X=(positions, measured) -> Y=(true amps)
##############################################################################

def build_amplitude_regressor(n_inputs=6, n_outputs=3):
    """
    A simple feedforward network that maps:
      (p1, p2, p3, p4, m1, m2, m3, m4) => (A1, A2, A3, A4).
    """
    inputs = layers.Input(shape=(n_inputs,))
    x = layers.Dense(64, activation='relu')(inputs)
    x = layers.Dense(64, activation='relu')(x)
    x = layers.Dense(64, activation='relu')(x)
    outputs = layers.Dense(n_outputs, activation='relu')(x)  # or 'softplus'
    model = models.Model(inputs, outputs, name="AmplitudeRegressor")
    return model
```

```python
# 1) Generate data
X_all, Y_all = generate_data(n_samples=5000, seed=42)


print("X_all shape:", X_all.shape)  # (5000, 8)
print("Y_all shape:", Y_all.shape)  # (5000, 4)

# 2) Split into train & val
n_train = 4000
X_train, X_val = X_all[:n_train], X_all[n_train:]
Y_train, Y_val = Y_all[:n_train], Y_all[n_train:]

# 3) Build the model
model = build_amplitude_regressor(n_inputs=6, n_outputs=3)
model.compile(optimizer='adam', loss='mse')

# 4) Train
history = model.fit(
    X_train, Y_train,
    validation_split=0.2,
    epochs=150,
    batch_size=32
)
```

```python
# 5) Evaluate or test
# For demonstration, let's see how well it does on some samples

X_all, Y_all = generate_data(n_samples=1, seed=32)
Y_pred = model.predict(X_val[:4])

# Print
print("True amps vs. Predicted amps (first 5 val samples):")
for i in range(4):
    print(f"Sample {i}")
    print("  True:", Y_val[i])
    print("  Pred:", Y_pred[i])

# Let's plot the training/validation loss
plt.plot(history.history["loss"], label="train_loss")
plt.plot(history.history["val_loss"], label="val_loss")
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.title("Training History")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from DeepPeak.signals import generate_gaussian_dataset
from DeepPeak.utils.visualization import plot_training_history, SignalPlotter
```

```python
SEQUENCE_LENGTH = 200
MAX_PEAKS = 3

def plot_sample(positions, measured, true_amp, sigma, signal=None):
    """
    Plot the full signal along with the pulse positions, measured amplitudes, and true amplitudes.
    
    Parameters:
      positions: 1D numpy array of pulse positions (padded with zeros if fewer than max pulses).
      measured: 1D numpy array of measured amplitudes at the pulse positions (same length as positions).
      true_amp: 1D numpy array of true amplitudes (same length as positions).
      sigma:     Standard deviation (width) of the Gaussian pulses.
      sequence_length: Number of points to use for the full signal (e.g., 200).
    """
    # Create a time vector spanning [0, 1]
    t = np.linspace(0, 1, SEQUENCE_LENGTH)
   
    
    # Identify valid pulses (those with a nonzero true amplitude)
    # (This works because in our generation function valid pulses have true_amp > 0;
    # padded pulses have true_amp == 0.)
    valid = true_amp > 0
    valid_positions = positions[valid]
    valid_true = true_amp[valid]
    valid_measured = measured[valid]

    # Create the plot
    plt.figure(figsize=(10, 6))


    if signal is not None: 
        plt.plot(t, signal, color='black', linestyle='--', linewidth=1)

    
    # Compute the full signal as the sum of all Gaussian pulses using the true amplitudes.
    # (Each pulse is given by: amplitude * exp(-((t - pos)^2)/(2*sigma^2)) )
    signal = np.zeros_like(t)
    for pos, amp in zip(valid_positions, valid_true):
        signal += amp * np.exp(-((t - pos) ** 2) / (2 * sigma ** 2))
        
    plt.plot(t, signal, label='Signal (sum of Gaussians)', color='blue')
    
    
    # Draw vertical dashed lines at each pulse position to indicate where the pulses occur.
    for pos in valid_positions:
        plt.axvline(x=pos, color='gray', linestyle='--', alpha=0.5)
    
    # Plot the measured amplitudes at the pulse positions (these are the values that your sensor sees,
    # which are affected by overlapping contributions).
    plt.scatter(valid_positions, valid_measured, color='red', marker='o', s=100, label='Measured amplitude')
    
    # Plot the true amplitudes (the actual amplitude of the individual pulses).
    plt.scatter(valid_positions, valid_true, color='green', marker='x', s=100, label='True amplitude')
    
    plt.xlabel("Normalized Time")
    plt.ylabel("Amplitude")
    plt.title("Signal with Gaussian Pulses: True & Measured Amplitudes and Positions")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def generate_sample(max_pulses, width):
    """
    Generate one synthetic sample.
    
    Each sample is a signal (of “sequence_length” 200, but here we only use the pulse locations)
    composed of a random number (0 to max_pulses) of Gaussian pulses.
    For each pulse:
      - A random position in [0,1] is chosen.
      - A true amplitude is chosen (here uniform between 0.5 and 1.5).
    
    The measured amplitude at the center of each pulse is computed as the sum of the contributions
    of all pulses at that position:
    
        measured[j] = sum_i true_amplitude[i] * exp(- ( (pos_j - pos_i)**2 / (2*sigma**2) ) )
    
    For samples with fewer than max_pulses pulses, the returned arrays are padded with zeros.
    
    Returns:
        padded_positions: numpy array of shape (max_pulses,)
        padded_measured: numpy array of shape (max_pulses,)
        padded_true: numpy array of shape (max_pulses,)
    """
    # Randomly choose the number of pulses (can be 0 to max_pulses)
    npulses = np.random.randint(0, max_pulses + 1)
    
    positions_list = []
    amplitudes_list = []
    
    if npulses > 0:
        for _ in range(npulses):
            pos = np.random.rand()  # random position in [0,1]
            amp = np.random.uniform(1, 10)  # true amplitude
            positions_list.append(pos)
            amplitudes_list.append(amp)
        positions_list = np.array(positions_list)
        amplitudes_list = np.array(amplitudes_list)

        sort_idx = np.argsort(positions_list)
        positions = positions_list[sort_idx]
        true_amplitudes = amplitudes_list[sort_idx]
    else:
        positions = np.array([])
        true_amplitudes = np.array([])
    
    # Compute measured amplitudes at each pulse's position,
    # taking into account the contributions (overlap) of all pulses.
    measured = []
    for j in range(npulses):
        m = 0.0
        for i in range(npulses):
            m += true_amplitudes[i] * np.exp(-((positions[j] - positions[i]) ** 2) / (2 * width ** 2))
        measured.append(m)
    measured = np.array(measured)
    
    # Pad arrays to length max_pulses (using 0 for "missing" pulses)
    padded_positions = np.zeros(max_pulses)
    padded_measured = np.zeros(max_pulses)
    padded_true = np.zeros(max_pulses)
    if npulses > 0:
        padded_positions[:npulses] = positions
        padded_measured[:npulses] = measured
        padded_true[:npulses] = true_amplitudes
        
    return padded_positions, padded_measured, padded_true
```

```python
pos, meas, true_amp = generate_sample(max_pulses=MAX_PEAKS, width=0.05)
plot_sample(pos, meas, true_amp, sigma=0.05)

X, y, signals, positions, true_amplitudes, measured = generate_dataset_DP(n_samples=1, max_peaks=MAX_PEAKS, width=0.02)
plot_sample(positions=positions, measured=measured, true_amp=true_amplitudes, signal=signals.squeeze(), sigma=0.05)
```

```python
def generate_dataset(n_samples, max_peaks, width):
    """
    Generate a dataset of n_samples.
    
    For each sample, the input is constructed by concatenating the pulse positions
    and the corresponding measured amplitudes (length 8 in total, for max_pulses=4).
    
    The target is the vector of true pulse amplitudes (length 4).
    """
    X = []
    y = []
    for _ in range(n_samples):
        pos, meas, true_amp = generate_sample(max_peaks, width)
        sample_input = np.concatenate([pos, meas])  # shape: (8,)
        X.append(sample_input)
        y.append(true_amp)  # shape: (4,)
    return np.array(X), np.array(y)



def compute_measured_amplitude(signal, x_values, peak_pos):
    nearest_idx = np.argmin(np.abs(x_values - peak_pos))
    return signal[nearest_idx]


def generate_dataset_DP(n_samples, max_peaks, width):
    signals, true_amplitudes, positions, _, x_values, num_peaks = generate_gaussian_dataset(
        n_samples=n_samples,
        sequence_length=SEQUENCE_LENGTH,
        n_peaks=(0, max_peaks),
        amplitude=(1, 60),
        position=(0.0, 1.0),
        width=(width, width),
        noise_std=0.01,
        categorical_peak_count=False,
    )
    
    X = []
    y = []
    measured_amplitudes = []
    for i in range(n_samples):
        signal = signals[i]               
        pos = positions[i]               
        amps = true_amplitudes[i]         
        n_peaks = int(num_peaks[i])
        measured = np.zeros(max_peaks)

        for j in range(n_peaks):
            measured[j] = compute_measured_amplitude(signal, x_values, pos[j])

        measured_amplitudes.append(measured)

        sample_input = np.concatenate([pos, measured])
        X.append(sample_input)
        y.append(amps)
    return np.array(X), np.array(y), signals, positions, true_amplitudes, np.array(measured_amplitudes)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

def build_model(input_dim, output_dim):
    model = keras.Sequential([
        # Increase neurons and add Batch Normalization + Dropout
        layers.Dense(128, activation='relu', input_shape=(input_dim,)),
        layers.BatchNormalization(),
        layers.Dropout(0.1),
        
        layers.Dense(128, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.1),
        
        # Optionally add another hidden layer
        layers.Dense(64, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.1),
        
        layers.Dense(output_dim, activation='linear')
    ])
    
    # You might also try tuning the learning rate
    optimizer = keras.optimizers.Adam(learning_rate=1e-3)
    model.compile(optimizer=optimizer, loss='mse')
    return model
```

```python
lr_scheduler = keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, verbose=1)

X_train, y_train, _, _, _, _ = generate_dataset_DP(
    n_samples=20_000, 
    max_peaks=MAX_PEAKS, 
    width=0.02
)

model = build_model(input_dim=2 * MAX_PEAKS, output_dim=MAX_PEAKS)

history = model.fit(
    X_train, y_train,
    epochs=40,
    batch_size=64,
    validation_split=0.2,
    verbose=1,
    callbacks=[lr_scheduler]
)

plot_training_history(history)
```

```python
X_val, y_val, _, _, _, _ = generate_dataset_DP(
    n_samples=1000,
    max_peaks=MAX_PEAKS, 
    width=0.05
)

y_pred = model.predict(X_val)

# Collect all valid predictions (ignore padded pulses, where true amplitude is 0)
true_vals = []
pred_vals = []
for sample_true, sample_pred in zip(y_val, y_pred):
    for true_amp, pred_amp in zip(sample_true, sample_pred):
        if true_amp > 0:  # if the pulse is present
            true_vals.append(true_amp)
            pred_vals.append(pred_amp)
            
true_vals = np.array(true_vals)
pred_vals = np.array(pred_vals)

plt.figure(figsize=(6, 6))
plt.scatter(true_vals, pred_vals, alpha=0.5, label='Pulses')
# Plot the identity line for reference
lims = [min(true_vals.min(), pred_vals.min()) - 0.1, max(true_vals.max(), pred_vals.max()) + 0.1]
plt.plot(lims, lims, 'r--', label='Ideal')
plt.xlabel("True Amplitude")
plt.ylabel("Predicted Amplitude")
plt.title("True vs. Predicted Pulse Amplitudes")
plt.legend()
plt.xlim(lims)
plt.ylim(lims)
plt.grid(True)
plt.show()
```
